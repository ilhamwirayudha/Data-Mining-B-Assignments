## **Part A – Data Preprocessing**

### 1. Load Dataset & Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv'
df = pd.read_csv(url)
```

### 3. Handle Missing Values

```python
# convert the price_in_rp column to a numeric, turning any 'N/A' into a (NaN) because price is a critical variable that cant be guessed
# next apply deletion, removing rows where price is now missing to ensure the validity of the analysis

df['price_in_rp'] = pd.to_numeric(df['price_in_rp'], errors='coerce')

price_missing_count = df['price_in_rp'].isnull().sum()
if price_missing_count > 0:
    df.dropna(subset=['price_in_rp'], inplace=True)
    print(f"Removed {price_missing_count} rows with missing price information.")
```

```python
# next I use imputation using central tendency
# for numerical columns, I filled missing values with median, which is robust to the outliers I observed in the data
# for categorical columns, I use Mode as it is the most logical replacement

numerical_cols_with_na = ['bedrooms', 'bathrooms', 'carports']
for col in numerical_cols_with_na:
    df[col].fillna(df[col].median(), inplace=True)

categorical_cols_with_na = ['property_condition', 'furnishing']
for col in categorical_cols_with_na:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df.isnull().sum())
```

### 4. Detect & handle outliers

```python
# I used boxplots to visually identify outliers in key numerical columns
# this provides a clear justification for our decision to keep them
# these extreme values likely represent legitimate luxury properties
# removing them would result in an incomplete and biased view of the market

plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['price_in_rp'])
plt.title('Boxplot of Price (in RP)')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['building_size_m2'])
plt.title('Boxplot of Building Size (m²)')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['land_size_m2'])
plt.title('Boxplot of Land Size (m²)')

plt.show()
```

### 5. Feature Engineering

```python
# I used the binning technique to convert the continuous price data into four discrete categories
# this feature simplifies the data and makes it easier to use for categorical analysis and visualization

price_bins = [0, 1e9, 3e9, 7e9, np.inf]
price_labels = ['Affordable', 'Mid-Range', 'Expensive', 'Luxury']
df['PriceCategory'] = pd.cut(df['price_in_rp'], bins=price_bins, labels=price_labels, right=False)
print(df['PriceCategory'].value_counts())
```

## **Part B – Exploratory Data Analysis (EDA)**

### Analysis 1: Distribution of House Prices

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='PriceCategory', order=['Affordable', 'Mid-Range', 'Expensive', 'Luxury'])
plt.title('Number of Properties in Each Price Category')
plt.xlabel('Price Category')
plt.ylabel('Number of Properties')
plt.show()

print("This bar plot clearly shows that the vast majority of properties fall into the 'Mid-Range'")
print("and 'Affordable' categories, confirming the data is concentrated at lower price points")
print("visualizes the distribution of properties across the price categories we created earlier") 
print("the bar plot is an effective alternative to a histogram for showing the concentration of data")
print ("demonstrating how many properties exist in each price segment")
```

### Analysis 2: Number of Properties per City

```python
plt.figure(figsize=(12, 7))
sns.countplot(y=df['city'])
plt.title('Number of Property Listings per city')
plt.show()

print("South Jakarta and Bogor have the most property listings, showing the geographical focus of the dataset")
print("the bar chart displays the |number of property listings in each area, highlighting where most properties are located")
print("this indicates that these areas are likely more active in the real estate market")
```

### Analysis 3: Relationship between Price and Building Size


```python
plt.figure(figsize=(8, 8))
df['furnishing'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'green', 'yellow'])
plt.title('Proportion of Property Furnishing Status')
plt.ylabel('')
plt.show()

print("This pie chart reveals majority of properties in the dataset over 69% are sold unfurnished")
print("the pie chart is used to show the proportional breakdown of a categorical variable, it effectively illustrates the percentage")
print("of homes that are furnished versus unfurnished, providing a clear and simple insight into the composition")
```

### Analysis 4: Price Distribution Across Different Cities

```python
plt.figure(figsize=(12, 7))
avg_price_by_city = df.groupby('city')['price_in_rp'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price_by_city.values, y=avg_price_by_city.index)
plt.title('Average Property Price by City')
plt.xlabel('Average Price (in Billions of Rupiah)')
plt.ylabel('City')
plt.ticklabel_format(style='sci', axis='x', scilimits=(9,9))
plt.show()

print("This bar plot of average prices clearly shows that South Jakarta is, on average")
print("the most expensive city for property, while Depok and Bogor are comparatively more affordable")
print("This is an effective way to analyze how a numerical variable differs across various categories")
print("the bar plotcompares a numerical value (average price) across different categories (cities)")
print("by first grouping the data to calculate the mean price for each city")
```

### Analysis 5: Correlation Heatmap of Numerical Features

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df.select_dtypes(include=np.number).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

print("This heatmap visually confirms that price is strongly correlated with building size, land size, and number of bathrooms")
print("the heatmap provides a quick, comprehensive overview of the linear relationships between all numerical variables in the dataset")
print("the colors make it easy to instantly spot strong positive or negative correlations")
```
